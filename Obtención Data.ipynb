### **Obtención desde PDF**

```python
!pip install PyPDF2
```

```python
import requests, PyPDF2
from io import BytesIO

url = 'https://servicios.adifse.com.ar/doc/GCRI01.pdf'
response = requests.get(url)
my_raw_data = response.content

with BytesIO(my_raw_data) as data:
    read_pdf = PyPDF2.PdfReader(data)

    for page in range(len(read_pdf.pages)):
        print(read_pdf.pages[page].extract_text())
```

### **Obtención desde Scrapping**

```python
from bs4 import BeautifulSoup
import requests
```

```python
page="https://www.lanacion.com.ar/"
r = requests.get(page.replace("\n",""))
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
links = soup.findAll('h2', {'class': 'com-title --xs'})

for link in links:
    print (link.find(class_="com-link").get_text(strip=True))
```

```python
import speech_recognition as sr
import IPython.display as ipd

!wget https://github.com/LCaravaggio/NLP/raw/main/01_obtenci%C3%B3n/AddresstotheWomenofAmerica.wav
```

```python
r = sr.Recognizer()
with sr.AudioFile('/content/AddresstotheWomenofAmerica.wav') as source:
    audio = r.record(source)

ipd.Audio('/content/AddresstotheWomenofAmerica.wav')

text=r.recognize_google(audio, show_all=True,language='en_US')

text
```
